```python
dup_mask = df.duplicated()
dup_rows = df[dup_mask]
print(dup_rows)
```

```python
# Get numeric and categorical column names
numeric_cols = df.select_dtypes(include=["int64", "float64"]).columns
cat_cols     = df.select_dtypes(include=["object"]).columns

print("Numeric:", list(numeric_cols))
print("Categorical:", list(cat_cols))
```

```python
from sklearn.impute import SimpleImputer
num_imputer = SimpleImputer(strategy='median')

df[numeric_cols] = num_imputer.fit_transform(df[numeric_cols])
```

```python
cat_imputer = SimpleImputer(strategy='most_frequent')

df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
df.isnull().sum()
```

```python
print(numeric_cols)
print(cat_cols)
```

```python
outlier_mask = pd.DataFrame(False, index=df.index, columns=numeric_cols)

for col in numeric_cols:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outlier_mask[col] = (df[col] < lower) | (df[col] > upper)

rows_with_any_outlier = df[outlier_mask.any(axis=1)]
```

```python
target = "SalePrice"
num_cols = numeric_cols.drop(target)

corr = df[numeric_cols.join(pd.Index([target]))].corr()[target].drop(target)
corr_sorted = corr.abs().sort_values(ascending=False)
print(corr_sorted.head(20))
```

```python
top_features = corr_sorted.head(10).index.tolist()
print(top_features)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for col in top_features:
    plt.figure(figsize=(5,4))
    sns.scatterplot(data=df, x=col, y=target, alpha=0.4)
    plt.title(f"{col} vs {target}")
    plt.show()
```

```python
import numpy as np
df["log_SalePrice"] = np.log1p(df["SalePrice"])
sns.scatterplot(data=df, x="Gr Liv Area", y="log_SalePrice")
```

```python
X = df.drop(columns=["SalePrice"])
y = df["SalePrice"]

X = pd.get_dummies(X, drop_first=True)  # one‑hot encode categoricals
```

```python
from sklearn.model_selection import cross_val_score, KFold
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

kf = KFold(n_splits=5, shuffle=True, random_state=42)

models = {
    "linreg": LinearRegression(),
    "rf": RandomForestRegressor(random_state=42)
}

for name, model in models.items():
    scores = cross_val_score(model, X, y, cv=kf, scoring="neg_root_mean_squared_error")
    print(name, -scores.mean())
```

```python
from sklearn.model_selection import RandomizedSearchCV
import numpy as np


param_dist = {
    'n_estimators': [100, 200, 500],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

rf = RandomForestRegressor(random_state=42)

rf_random = RandomizedSearchCV(estimator=rf, param_distributions=param_dist,
                               n_iter=10, cv=5, verbose=2, random_state=42, n_jobs=-1,
                               scoring='neg_root_mean_squared_error')

rf_random.fit(X, y)

print("Best Parameters found:", rf_random.best_params_)
print("Best RMSE score:", -rf_random.best_score_)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

y_pred = models['rf'].fit(X, y).predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x=y, y=y_pred, alpha=0.5)
plt.plot([y.min(), y.max()], [y.min(), y.max()], color='red', lw=2) # Diagonal line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Random Forest: Actual vs Predicted')
plt.show()
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y, y_pred)
print(f"Your model explains {r2*100:.2f}% of the data variance.")
```
